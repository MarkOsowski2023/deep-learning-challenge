## Preprocessing

```python
# sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# tensorflow
import tensorflow as tf

# tuner
import keras_tuner as kt

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Data/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

df = application_df.drop(columns=['EIN', 'NAME'], axis=1)
df.head()
```

```python
# Determine the number of unique values in each column.
print(df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
print(df['APPLICATION_TYPE'].value_counts())
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# variable name `application_types_to_replace`
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']
# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
print(df['CLASSIFICATION'].value_counts())
```

```python
# CLASSIFICATION value counts

classification_counts = df['CLASSIFICATION'].value_counts()

classification_counts.loc[classification_counts > 1]
```

```python
# cutoff for classifications: value counts < 1883
# create variable classifications_to_repalce

classifications_to_replace = classification_counts.loc[classification_counts < 1883].index.tolist()

print(classifications_to_replace)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")
    
# Check
df['CLASSIFICATION'].value_counts()
```

```python
# check data types for columns
# if dtype is object it is categorical data
df.info()
```

```python
# create list of categorical features
categorical_cols = [col for col in df.columns if df.dtypes[col] == 'object']
print(categorical_cols)
```

```python
# Convert categorical features to numeric with pd.get_dummies
df_with_dummies = pd.get_dummies(df, columns=categorical_cols)
df_with_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
features = df_with_dummies.drop(columns='IS_SUCCESSFUL', axis=1)
target = df_with_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42, stratify=target)
```

```python
# Creating Scaler
# scaler
scaler = StandardScaler()

# fit 
X_scaler = scaler.fit(X_train)

# transform
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=9, activation="relu", input_dim=43))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=9, activation="relu"))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
nn_fit = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Tuning Model
# creating function that creates a new Swquential with Hyperparameter options

def create_nn_model(hp):
    nn_model = tf.keras.models.Sequential()
    
    # allowing tuner to decide which activation to use in the first layer
    activation = hp.Choice("activation",["relu", "tanh"])
    
    # allow tuner to decide number of neurons
    nn_model.add(tf.keras.layers.Dense(units=hp.Int("first_units",
                                                   min_value = 1,
                                                   max_value = 30,
                                                   step=5), activation=activation, input_dim=43))
    
    # allow tuner to decide number of hidden layers and number of neurons in the hidden layer
    # for loop
    for i in range(hp.Int("num_layers", 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int("units_" + str(i),
                                                       min_value=1,
                                                       max_value=30,
                                                       step=5), activation=activation))
        
    # output layer
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
    
    # compile
    nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    
    # return model
    return nn_model
        
    
```

```python
# tuner
tuner = kt.Hyperband(create_nn_model,
                    objective="val_accuracy",
                    max_epochs=100,
                    hyperband_iterations=2)
```

```python
# tuner search
tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
```

```python
top_model = tuner.get_best_models(1)
```
